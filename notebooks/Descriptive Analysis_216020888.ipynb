```python
import pandas as pd
import numpy as np
import matplotlib 

data = {
    'Provinces': ['WC', 'EC', 'NC', 'FS', 'KZN', 'NW', 'GP', 'MP', 'LP'],
    'Taxi (No in Thousands)': [968000, 1011000, 123000, 382000, 1848000, 585000, 3780000, 775000, 1240000],
    'Bus (No in Thousands)': [242000, 117000, 38000, 72000, 351000, 125000, 430000, 294000, 217000],
    'Train (No in Thousands)': [74000, 12000, 0, 0, 32000, 0, 181000, 4000, 0]
}

df = pd.DataFrame(data)

# Display the DataFrame
print("Original Data:")
print(df)

print("\nDescriptive Analysis:")

# Statistics Summary
print("\nSummary Statistics:")
print(df.describe())

# Total count of data
print("\nTotal Count:")
print(df.count())

# Total sum of data
print("\nTotal Sum:")
print(df.sum())

# Mean of data
print("\nMean:")
print(df.mean())

# Median of data
print("\nMedian:")
print(df.median())

# Mode of data
print("\nMode:")
print(df.mode().iloc[0])

# Maximum value of data
print("\nMaximum Value:")
print(df.max())

# Minimum value of data
print("\nMinimum Value:")
print(df.min())

# Range of data
print("\nRange:")
print(df.max() - df.min())

# Variance of data
print("\nVariance:")
print(df.var())

# Standard deviation of data
print("\nStandard Deviation:")
print(df.std())

# Correlation between columns
print("\nCorrelation:")
print(df.corr())

# Covariance between columns
print("\nCovariance:")
print(df.cov())
```
